```python
import pandas as pd
```

```python
df = pd.read_csv("DHL_Facilities.csv")
```

```python
df.head()
```

```python
# LOCATIONTY: The type of location (business or residential). (Text)
df["LOCATIONTY"].value_counts()
```

```python
# LOCATIONTH: Specifies whether the facility is a building or a street-based site. (Text)
df["LOCATIONTH"].value_counts()
```

```python
# Assuming we go with ["Staffed facility"] as our dataset
fdf = df[df["LOCATIONTY"] == "DHL Staffed Facility"]
```

```python
fdf["STATE"]
```

```python
fdf["NAME"].value_counts()
```

```python
REGIONS = {
    "NORTHEAST": ["CT", "ME", "MA", "NH", "RI", "VT", "NJ", "NY"],
    "MIDWEST": ["IL", "IN", "MI", "OH", "WI", "IA", "KS", "MN", "MO", "NE", "ND", "SD"],
    "SOUTH": ["TX", "OK", "AR", "LA", "KY", "TN", "AL", "MS", "FL", "GA", "SC", "NC", "VA", "WV", "DC", "MD", "DE"],
    "WEST": ["WA", "OR", "CA", "NV", "ID", "MT", "WY", "UT", "AZ", "NM", "CO"],
    "EAST": ["CT", "ME", "MA", "NH", "RI", "VT", "NJ", "NY", "IL", "IN", "MI", "OH", "WI", "IA", "KS", "MN", "MO", "NE", "ND", "SD", "TX", "OK", "AR", "LA", "KY", "TN", "AL", "MS", "FL", "GA", "SC", "NC", "VA", "WV", "DC", "MD", "DE"],
    "USAMAINLAND": ["CT", "ME", "MA", "NH", "RI", "VT", "NJ", "NY", "IL", "IN", "MI", "OH", "WI", "IA", "KS", "MN", "MO", "NE", "ND", "SD", "TX", "OK", "AR", "LA", "KY", "TN", "AL", "MS", "FL", "GA", "SC", "NC", "VA", "WV", "DC", "MD", "DE", "WA", "OR", "CA", "NV", "ID", "MT", "WY", "UT", "AZ", "NM", "CO"],
}

REGIONS_DFS = {reg: None for reg in REGIONS.keys()}

for reg, states in REGIONS.items():
    rdf = fdf[fdf["STATE"].isin(states)]
    REGIONS_DFS[reg] = rdf

for reg, rdf in REGIONS_DFS.items():
    # Reset the index to create a new 'region_index' column and keep the old index
    rdf.reset_index(drop=False, inplace=True)
    # Rename the old index to 'original_index' and the new index to 'region_index'
    rdf.rename(columns={"index": "original_index"}, inplace=True)
    # If 'level_0' appears, drop it
    if 'level_0' in rdf.columns:
        rdf.drop(columns=['level_0'], inplace=True)
    filename = reg + "facilities.csv"
    rdf.to_csv(filename)
    
```

```python
import folium

lat = "LATITUDE"
long = "LONGITUDE"
# Create a base map
m = folium.Map(location=[fdf[lat].mean(), fdf[long].mean()], zoom_start=4.48)

# Add points to the map
for _, row in fdf.iterrows():
    folium.CircleMarker(
        location=[row[lat], row[long]],
        radius=3,  # Size of the circle
        color='black',  # Outline color
        fill=True,
        fill_color='black',  # Fill color
        fill_opacity=0.6,
    ).add_to(m)

# Save the map to an HTML file (LATER maybe)
# m.save('map.html')
```
